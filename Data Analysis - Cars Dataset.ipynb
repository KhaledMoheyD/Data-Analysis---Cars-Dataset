### Data Cleaning

#### finding all records with null values

```python
c_df[c_df.isnull().any(axis=1)]
```

#### drop records with missing values

```python
c_df.dropna(axis=0,how='all', inplace=True)
```

```python
c_df[c_df.isnull().any(axis=1)]
```

#### fill missing values

```python
c_df.Cylinders.unique()
```

```python
c_df['Cylinders'].fillna(c_df['Cylinders'].mode()[0], inplace=True)
```

```python
c_df.head()
```

#### Converting MSRP and Invoice into numerical data 

```python
#### remove characters from MSRP
temp = pd.DataFrame(columns=['0','1'])
temp[['0','1']] = c_df['MSRP'].str.split('$',expand=True)
temp[['0','1']] = temp['1'].str.split(',',expand=True)
c_df['MSRP'] = temp['0'] + temp['1']
c_df['MSRP'].head()
```

```python
#### Remove characters from Invoice
temp[['0','1']] = c_df['Invoice'].str.split('$',expand=True)
temp[['0','1']] = temp['1'].str.split(',',expand=True)
c_df['Invoice'] = temp['0'] + temp['1']
c_df['Invoice'].head()
```

```python
### convert categorical to numerical in 'MSRP' , 'Invoice'
c_df = c_df.astype({'MSRP':'float','Invoice':'float'})
c_df
```
